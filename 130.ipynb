```python
Given a 2D board containing 'X' and 'O' (the letter O), capture all regions surrounded by 'X'.

A region is captured by flipping all 'O's into 'X's in that surrounded region.

Example:

X X X X
X O O X
X X O X
X O X X
After running your function, the board should be:

X X X X
X X X X
X X X X
X O X X
Explanation:

Surrounded regions shouldn’t be on the border, which means that any 'O' on the border of the board are not flipped to 'X'. Any 'O' that is not on the border and it is not connected to an 'O' on the border will be flipped to 'X'. Two cells are connected if they are adjacent cells connected horizontally or vertically.

来源：力扣（LeetCode）
链接：https://leetcode-cn.com/problems/surrounded-regions
著作权归领扣网络所有。商业转载请联系官方授权，非商业转载请注明出处。
```

```python
from typing import List
class Solution:
    def solve(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        if not board:
            return 
        m=len(board)
        n=len(board[0])
        root=[['X']*n for _ in range(m)]
    
        def checkSurround(board,pos,root,m,n):
            x,y=pos
            if x < m and x>-1 and  y < n and y>-1 and board[x][y]=='O' and root[x][y]!='O':
                root[x][y]='O'
                checkSurround(board,(x-1,y),root,m,n)
                checkSurround(board,(x+1,y),root,m,n)
                checkSurround(board,(x,y-1),root,m,n)
                checkSurround(board,(x,y+1),root,m,n)

        for i,row in enumerate(board):
            for j,ele in enumerate(row):
                if (i==0 or i==m-1 or j==0 or j==n-1) and ele=='O':
                    checkSurround(board,(i,j),root,m,n)
        for i in range(m):
            for j in range(n):
                if root[i][j]!=board[i][j]:
                    board[i][j]=root[i][j]

example=[
        ['X','O','X','X'],
        ['X','O','O','X'],
        ['X','X','O','X'],
        ['X','O','X','X'],
        ['X','X','X','X']
    ]

Solution().solve(example)
print(example)
```

```python
#BFS搜索
from typing import List
class Solution:
    def solve(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        if not board:
            return 
        m=len(board)
        n=len(board[0])
    
        def checkSurround(board,pos,m,n):
            from collections import deque
            queue=deque()
            queue.appendleft(pos)
            while queue:
                i,j=queue.pop()
                if 0<=i<m and 0<=j<n and board[i][j]=='O':
                    board[i][j]='#'
                    for x,y in [(-1,0),(1,0),(0,-1),(0,1)]:
                        queue.appendleft((i+x,j+y))

        for i,row in enumerate(board):
            for j,ele in enumerate(row):
                if (i==0 or i==m-1 or j==0 or j==n-1) and ele=='O':
                    checkSurround(board,(i,j),m,n)
        for i in range(m):
            for j in range(n):
                if board[i][j]=='O':
                    board[i][j]='X'
                elif board[i][j]=='#':
                    board[i][j]='O'

example=[
        ['X','O','X','X'],
        ['X','O','O','X'],
        ['X','X','O','X'],
        ['X','O','X','X'],
        ['X','X','X','X']
    ]

Solution().solve(example)
print(example)
```
