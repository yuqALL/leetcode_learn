```python
魔咒串由许多魔咒字符组成，魔咒字符可以用数字表示。例如可以将魔咒字符 1、2 拼凑起来形成一个魔咒串[1,2] 。

一个魔咒串 S 的非空子串被称为魔咒串 S 的生成魔咒。

例如 S=[1,2,1] 时，它的生成魔咒有[1]、[2]、[1,2]、[2,1]、[1,2,1] 五种。S=[1,1,1] 时，它的生成魔咒有[1]、[1,1]、[1,1,1] 三种。

最初 S 为空串。共进行 n 次操作，每次操作是在 S 的结尾加入一个魔咒字符。每次操作后都需要求出，当前的魔咒串 S 共有多少种生成魔咒。

输入：
第一行一个整数 n 。
第二行 n 个数，第 i 个数表示第 i 次操作加入的魔咒字符。

输出：
输出 n 行，每行一个数。第 i 行的数表示第 i 次操作后 S 的生成魔咒数量。

示例：
输入：
7
1 2 3 3 3 1 2

输出：
1
3
6
9
12
17
22

数据范围与提示：
对于 10% 的数据，1 ≤ n ≤ 10；
对于 30% 的数据，1 ≤ n ≤ 100；
对于 60% 的数据，1 ≤ n ≤ 1000；
对于 100% 的数据，1 ≤ n ≤ 1000000。

用来表示魔咒字符的数字 x 满足 1 ≤ x ≤ 10^9。
```

```python
class SamNode:
    def __init__(self):
        self.len=0
        self.link=0
        self.next={}

class Sam:
    def __init__(self):
        self.st=[SamNode()]
        self.st[0].len=0
        self.st[0].link=-1
        self.last=0
        self.size=1

    def extend(self,c):
        cur=self.size
        self.st.append(SamNode())
        self.st[cur].len=self.st[self.last].len+1
        self.size+=1
        p=self.last
        while p!=-1 and c not in self.st[p].next:
            self.st[p].next[c]=cur
            p=self.st[p].link
            
        if p==-1:
            self.st[cur].link=0
        else:
            q=self.st[p].next[c]
            if self.st[q].len==self.st[p].len+1:
                self.st[cur].link=q
            else:
                clone=self.size
                self.size+=1
                self.st.append(SamNode())
                self.st[clone].len=self.st[p].len+1
                self.st[clone].link=self.st[q].link
                self.st[clone].next=self.st[q].next.copy()
                while p!=-1 and self.st[p].next[c] == q:
                    self.st[p].next[c]=clone
                    p=self.st[p].link
                self.st[q].link=self.st[cur].link=clone

        self.last=cur
    
    def search_suffix(self,c):
      
        pass

    def num_substrings_next(self,cur=0):
        if not self.st[cur].next:
            return 0
        size=0
        for k in self.st[cur].next:
            size+=self.num_substrings_next(self.st[cur].next[k])+1
     
        return size
    
    def num_substrings_link(self):
        s=0
        cur=1
        while cur<self.size:
            s+=self.st[cur].len-self.st[self.st[cur].link].len
            cur+=1
        return s
    
    def total_substrings_len_next(self,cur=0,ans=1):
        if not self.st[cur].next:
            return 0
        size=0
        for k in self.st[cur].next:
            # print("pre",k,self.st[cur].next,size,ans)
            size+=self.total_substrings_len_next(self.st[cur].next[k],ans+1)+ans
            # print(k,self.st[cur].next,size,ans)
        return size
    
    def total_substrings_len_link(self):
        s=0
        cur=1
        while cur<self.size:
            lc=self.st[cur].len
            lcl=self.st[self.st[cur].link].len
            s+=(lc*(lc+1)-lcl*(lcl+1))//2
            # print(cur,self.st[cur].link,lc,lcl)
            cur+=1
        return s


class Solution:
    def solveCharms(self,n,nums):
        sam=Sam()
        s=0
        res=[]
        res_link=[]
        res_len_next=[]
        res_len_link=[]
        for i in range(n):
            sam.extend(nums[i])
            res.append(sam.num_substrings_next())
            res_link.append(sam.num_substrings_link())
            # print(i)
            res_len_next.append(sam.total_substrings_len_next())
            res_len_link.append(sam.total_substrings_len_link())
        print("num next",res)
        print("num link",res_link)
        print("length next",res_len_next)
        print("length link",res_len_link)
        return res

print(Solution().solveCharms(6,[1, 2, 3, 3, 3, 1]))
print(Solution().solveCharms(3,[1, 2, 1]))
print(Solution().solveCharms(5,['a', 'b', 'c','b','c']))
print(Solution().solveCharms(10,[1,2,3,1,1, 1, 4, 1, 3, 5]))
```
