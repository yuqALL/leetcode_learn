```python
#大根堆
class HeapSort:
    def __init__(self,nums):
        self.n=len(nums)
        self.nums=nums
        print("pre",self.nums)
        for i in range(self.n//2-1,-1,-1):
            self.heapify(i)
        print("build",self.nums)

    def heapify(self,i):
        ri=(i+1)*2
        le=ri-1
        largest=i
        if le<self.n and self.nums[le] > self.nums[i]:
            largest=le
        if ri<self.n and self.nums[ri]>self.nums[largest]:
            largest=ri
        if largest!=i:
            t=self.nums[i]
            self.nums[i]=self.nums[largest]
            self.nums[largest]=t
            self.heapify(largest)
    
    def sort(self):
        for i in range(self.n-1,0,-1):
            t=self.nums[i]
            self.nums[i]=self.nums[0]
            self.nums[0]=t
            self.n-=1
            self.heapify(0)
        print("sort",self.nums)

nums=[57, 40, 38, 11, 13, 34, 48, 75, 6, 19, 9, 7]

heapsort=HeapSort(nums)
heapsort.sort()
print(id(nums),id(heapsort.nums))
```
