Shell 排序是以它的发明者命名的，也称为缩小增量排序法。Shell 排序对不相邻的记录进行比较和移动：

1.  将待排序序列分为若干子序列（每个子序列的元素在原始数组中间距相同）
2.  对这些子序列进行插入排序
3.  减小每个子序列中元素之间的间距，重复上述过程直至间距减少为 1

Shell 排序的复杂度和间距序列的选取（就是间距如何减小到 1）有关，比如“间距每次除以 3”的 Shell 排序的复杂度是 $O(n^{3/2})$ 。

```python
def insertion_sort(nums):
    for i in range(1,len(nums)):
        key=nums[i]
        j=i-1
        while j>-1 and nums[j]>key:
            nums[j+1]=nums[j]
            j-=1
        nums[j+1]=key

def shell_sort(nums):
    n=len(nums)
    if not n or n==1:
        return nums
    
    factor=2
    d=n//factor
    while d>1:
        temp=[]
        for i in range(d):
            cur=[nums[j] for j in range(i,n,d)]
            insertion_sort(cur)
            temp.extend(cur)
        nums=temp
        d=d//factor
    insertion_sort(nums)
    return nums

inputs=[99, 5, 69, 33, 56, 13, 22, 55, 77, 48, 12, 88, 2,69,99,2]
shell_sort(inputs)


```
